```python
import pandas as pd
import numpy
```

```python
df = pd.read_csv('/content/Transjakarta.csv',sep=',')
df
df[df['payCardSex'] == 'M']
```

```python
df.info()
```

**ADD GENERASI**

```python
def gencek(collection):
    if (collection['payCardBirthDate'] >= 1946) & (collection['payCardBirthDate'] < 1964):
      generasi = 'Baby Boomers'
    elif (collection['payCardBirthDate'] >= 1964) & (collection['payCardBirthDate'] < 1980):
      generasi = 'Generasi X'
    elif (collection['payCardBirthDate'] >= 1980) & (collection['payCardBirthDate'] < 2000):
      generasi = 'Generasi Y'
    elif (collection['payCardBirthDate'] >= 2000):
      generasi = 'Generasi Z'
    else:
      generasi = 'Tidak ada Data'
    return generasi


df['Generasi'] = df.apply(gencek, axis=1)
df
```

**Perubahan payCard SEX**

```python
def ubahjeniskelamin(sex):
  if sex == 'F':
    return 'Female'
  elif sex == 'M':
    return 'Male'
  else:
    return 'Unknown'

df['payCardSex'] = df['payCardSex'].apply(ubahjeniskelamin)
df
```

**Jumlah Pemberhentian**

```python
def JumlahPemberhentian(collection):
    Jumlah = collection['stopEndSeq'] -collection['stopStartSeq']
    return Jumlah


df['JumlahPemberhentian'] = df.apply(JumlahPemberhentian, axis=1)
df
```
